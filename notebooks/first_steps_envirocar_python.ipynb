### Fetch track list

```python
import json
from urllib import urlopen
```

```python
#df_tracks = pd.read_json('https://envirocar.org/api/stable/tracks')
#df_tracks.head()
jsonurl = urlopen('https://envirocar.org/api/stable/tracks')
tracks = json.loads(jsonurl.read())
```

### Load all tracks and compare average GPS speed with ordinary speed for each track

```python
for value in tracks['tracks']:
    id = value['id']
    df = pd.read_csv('https://envirocar.org/api/stable/tracks/'+id+'.csv', sep=';\s+')
    #print df.dtypes
    
    gps_speed = -99
    speed = -99
    
    try:
        gps_speed = df['GPS Speed(km/h)'].mean()  
    except:
        pass
        
    try: 
        speed = df['Speed(km/h)'].mean()
    except:
        pass
        
    print 'GPS Speed: {0:.2f}, Speed: {1:.2f}'.format(gps_speed, speed)
    
```
